```python
import random
class foodorder:
    
    def __init__(self):
        self.admin_details={}
        self.foods={}
        self.food=[]
        self.main_menu()
        
        
    def admin_registration(self):
        self.admin_name=input("\n Enter your full name ")
        a=int(input("\n Enter your number , it must be 10 digits "))
        if len(str(a))==10:
            self.admin_contact_number=a
        else:
            print("\n Your mobail number does not contains 10 digits ")
        self.mail_id=input("\n Enter your mail-id ")
        self.password=input("Enter your password : ")
        s={}
        s["admin_name"]=self.admin_name
        s["admin_contact_number"]=self.admin_contact_number
        s["admin_mail_id"]=self.mail_id
        s["admin_password"]=self.password
        self.admin_details[self.admin_contact_number]=s
        print("\n Your registration process completed sucessfully ")
        print("\n admin_details \n\n ",self.admin_details)
        
        self.main_menu()
        
    def admin_login(self):
        num=int(input("\n Enter your contact number to loggin : "))
        if num in self.admin_details:
            pas=input("\n Enter your password")
            if self.admin_details[num]["admin_password"]==pas:
                print("\n Logged in sucessfully ")
                self.menu()
                
    def menu(self):
        x=int(input('''Enter\n 
        1 - create food item 
        2 - update food details
        3 - view food details
        4 - delete food item
        5 - Return to main menu'''))
        if x==1:
            self.create_food()
        elif x==2:
            self.update_food()
        elif x==3:
            self.view_food()
        elif x==4:
            self.delete_food()
        elif x==5:
            self.main_menu()
            
    def create_food(self):   
        self.food_id=random.randrange(1,100,1)
        print("Your system generated food id is :",self.food_id)
        n=input(" \n Enter the food Name ")
        self.food.append(n)
        p=int(input(" \n Enter the food  price "))
        self.food.append(p)
        q=int(input(" \n Enter the quantity of food "))
        self.food.append(q)
        self.foods[self.food_id]=self.food
        print(" \n  food saved ",self.foods)
        
        
        self.menu()
    
    def update_food(self):
        pid=input(" \n Enter your food id ")
        if pid in self.foods:
            a=int(input(''' \n  Enter \n 1 -  To change food id
            \n 2 - To change food name  \n 3- To change the food price \n 4 - To change food quantity'''))
            if a==1:
                b=self.foods[pid]
                self.foods.pop(pid)
                c=input("Enter new product id")
                self.foods[c]=b
                print("New id updated successfully")
            if a==2:
                self.foods[pid][0]=input("Please enter new name to the food")
                print(" \n  Your food name has been changed sucessfully ")
            if a==3:
                self.foods[pid][1]=input("please enter new price for food")
                print("\n Your food price is changed")
            if a==4:
                self.foods[pid][2]=input("Please enter new quantity for food ")
                print("\n New quantity is updated")
            else:
                print("\n your input is wrong ")
        else:
            print("\n Your food id not found")
            
        self.menu()
        
    def view_food(self):
        pid=input(" \n Enter your food id ")
        if pid in self.foods:
            a=int(input('''' \n  Enter \n 1 - To view food name \n 2 - To view price  \n 
                        3 - To view the quantity \n 4  - To see all details'''))
            if a==1: 
                print(" \n  Name of your food:  ",self.foods[pid][0])
            if a==2:
                print(" \n  Price of food is : ",self.foods[pid][1])
            if a==3:
                print(" \n  Quantity of food is : ",self.foods[pid][2])
            if a==4:
                print(" ")
                print(" Namae of your food :  ",self.foods[pid][0])
                print(" Price of food is  : ",self.foods[pid][1])
                print(" Quantity of food is : ",self.foods[pid][2])
        
        else:
            print("Your food id not found")
            
        self.menu()
    
    def delete_food(self):
        pid=input(" \n Enter your food id ")
        if pid in self.foods:
            inp=input(" \n  Do you want to delet this product if Yes type  Y (else) type  N")
            if inp=="Y" or "y":
                    self.foods.pop(pid)
                    print(" \n Your list was updated after removing that food ")
            if inp=="N" or "n":
                    pass
        else:
            
            print(" \n OOPS !!...Your food id was not found ")
            
        self.menu()
        
    def user_registration(self):
        pass
    def user_login(self):
        pass
        
    def exit(self):
        print("Thankyou for visisting us")
        
    
            
    def main_menu(self): 
        
        x=int(input('''Enter \n 1 - Admin Registration \n
        2 - Admin Login \n
        3 - User Registration \n
        4 - User Login
        5 - Exit'''))
        if x==1:
            self.admin_registration()
        elif x==2:
            self.admin_login()
        elif x==3:
            self.user_registration()
        elif x==4:
            self.user_login()
        elif x==5:
            self.exit()
        else:
            print("You had entered wrong input")
                      
```
